```python
import numpy as np
import scipy.sparse as sp
from scipy import io
import scipy.stats as stats
import scipy.linalg as lin
import matplotlib.pyplot as plt
import nmarkov as nm
```

```python
mat = io.loadmat('test.mat', squeeze_me=True)
```

```python
Q=mat['Q']
x=mat['x']
r=mat['r']
```

```python
r
```

```python
plt.spy(Q, markersize=0.1)
```

```python
ma = nm.nctmc(Q)
ma.params.rmax = 1000
```

```python
%%time
res=ma.trwd(t=np.linspace(0,100, num=10000), x0=x, rwd=r)
```

```python
plt.plot()
plt.plot(res['t'], res['irwd'])
```
