```python
class GitHub(object):
    def __init__(self, **config_options):
        self.__dict__.update(**config_options)
        self.session = requests.Session()
        if hasattr(self, 'api_token'):
           self.session.headers['Authorization'] = 'token {}'.format(self.api_token)
        elif hasattr(self, 'username') and hasattr(self, 'password'):
           self.session.auth = (self.username, self.password)

    def get(self, url):
        result = self.session.get(url)
        return json.loads(result.text)
```

```python
class Dir(object, ):
    def __init__(self, github, url):
        self.github = github
        data = self.github.get(url)
        if 'message' in data:
            raise RuntimeError(data)
        self.path = {x['path']: File(self.github, x['url']) if x['type'] == 'blob' else Dir(self.github, x['url']) for x in data['tree']}
    
class File(object):
    def __init__(self, github, url):
        self.github = github
        data = self.github.get(url)
        if 'message' in data:
            raise RuntimeError(data)
        if 'content' in data:
            self.content = data['content'] #base64.b64decode(data['content'].encode('UTF-8')).decode()
            self.lines = self.content.count('\n')
            self.target = True
        else:
            self.content = None
            self.target = False
```
